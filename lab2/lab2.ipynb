```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Создание пустого белого изображения
image = np.ones((500, 500, 3), dtype=np.uint8) * 255

# Определение точек многоугольника
# Пример: пятиугольник
points = np.array([[100, 100], [400, 100], [450, 250], [250, 400], [50, 250]])

# Рисование многоугольника без закрашивания (контур)
contour_color = (0, 0, 255)  # Красный цвет в BGR
contour_thickness = 2
cv2.polylines(image, [points], isClosed=True, color=contour_color, thickness=contour_thickness)

# Закрашивание многоугольника
fill_color = (0, 255, 0)  # Зеленый цвет в BGR
cv2.fillPoly(image, [points], color=fill_color)

# Дополнительный пример: закрашивание нескольких многоугольников

# Треугольник
triangle = np.array([[300, 300], [400, 300], [350, 400]])
cv2.fillPoly(image, [triangle], color=(255, 0, 0))  # Синий цвет

# Шестиугольник
hexagon = np.array([
    [200, 350], [250, 300], [350, 300],
    [400, 350], [350, 400], [250, 400]
])
cv2.fillPoly(image, [hexagon], color=(255, 255, 0))  # Желтый цвет

# Конвертация BGR в RGB для отображения с помощью matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Отображение изображения
plt.figure(figsize=(8, 8))
plt.imshow(image_rgb)
plt.title('Закрашивание многоугольников')
plt.axis('off')
plt.show()
```
