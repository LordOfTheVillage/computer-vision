```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Загрузка изображения
image_path = '../images/IMG_20130818_081052.jpg'
image = cv2.imread(image_path)

# Проверка успешности загрузки
if image is None:
    raise FileNotFoundError(f"Изображение по пути {image_path} не найдено.")

# Конвертация изображения в цветовую модель RGB для корректного отображения
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Построение гистограммы для цветных каналов (Красный, Зеленый, Синий)
plt.figure(figsize=(10, 5))
color_channels = ('red', 'green', 'blue')
for i, color in enumerate(color_channels):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color=color, label=f'{color.capitalize()} канал')
    plt.xlim([0, 256])
plt.title('Гистограмма цветовых каналов')
plt.xlabel('Интенсивность')
plt.ylabel('Количество пикселей')
plt.legend()
plt.show()

# Конвертация изображения в оттенки серого
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Построение гистограммы для оттенков серого
plt.figure(figsize=(10, 5))
histogram_gray = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(histogram_gray, color='gray')
plt.title('Гистограмма оттенков серого')
plt.xlabel('Интенсивность')
plt.ylabel('Количество пикселей')
plt.xlim([0, 256])
plt.show()
```
