```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#  
image_path = '../images/IMG_20130818_081052.jpg'
image = cv2.imread(image_path)

#   
if image is None:
    raise FileNotFoundError(f"   {image_path}  .")

#      RGB   
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#      (, , )
plt.figure(figsize=(10, 5))
color_channels = ('red', 'green', 'blue')
for i, color in enumerate(color_channels):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color=color, label=f'{color.capitalize()} ')
    plt.xlim([0, 256])
plt.title('  ')
plt.xlabel('')
plt.ylabel(' ')
plt.legend()
plt.show()

#     
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#     
plt.figure(figsize=(10, 5))
histogram_gray = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(histogram_gray, color='gray')
plt.title('  ')
plt.xlabel('')
plt.ylabel(' ')
plt.xlim([0, 256])
plt.show()
```
