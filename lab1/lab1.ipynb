```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Создание пустого белого изображения
image = np.ones((500, 500, 3), dtype=np.uint8) * 255

# Рисование отрезка
start_point = (50, 50)
end_point = (450, 50)
color = (255, 0, 0)  # Синий цвет в BGR
thickness = 2
cv2.line(image, start_point, end_point, color, thickness)

# Рисование окружности
center_coordinates = (250, 250)
radius = 100
color = (0, 255, 0)  # Зеленый цвет в BGR
thickness = 2
cv2.circle(image, center_coordinates, radius, color, thickness)

# Рисование эллипса
center_coordinates = (250, 250)
axes_lengths = (150, 75)
angle = 0
start_angle = 0
end_angle = 360
color = (0, 0, 255)  # Красный цвет в BGR
thickness = 2
cv2.ellipse(image, center_coordinates, axes_lengths, angle, start_angle, end_angle, color, thickness)

# Конвертация BGR в RGB для отображения с помощью matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Отображение изображения
plt.imshow(image_rgb)
plt.title('Рисование отрезков, окружности и эллипса')
plt.axis('off')
plt.show()
```
